```python
import pandas as pd
import plotly.express as px

timesData = pd.read_csv("/content/sample_data/timesData.csv")
timesData.head(10)
```

```python
df = timesData.iloc[:100, :]
df
```

```python
# Prepare data frame
df = timesData.iloc[:100, :]

fig = px.line(df, x="world_rank", y="citations", title="Citation vs World Rank of Top 100 Universities", markers=True)
fig.add_scatter(x=df["world_rank"], y=df["teaching"], mode="lines+markers", name="teaching", line=dict(color="rgba(80, 26, 80, 0.8)"))

fig.update_xaxes(title="World Rank")
fig.show()
```

```python
# Data preparation
df2016 = timesData[timesData.year == 2016].iloc[:20, :]
df2016['num_students'] = df2016['num_students'].str.replace(',', '.').astype(float)

# Create a scatter plot using Plotly Express
fig = px.scatter(
    df2016,
    x='world_rank',
    y='teaching',
    color='international',
    size='num_students',
    title='Teaching vs World Rank of Top 20 Universities in 2016',
    color_continuous_scale='Viridis',
    opacity=0.8,
render_mode='webgl',
)

fig.update_traces(marker=dict(showscale=True))


fig.show()
```

```python
import plotly.express as px

# Prepare data frame
df = timesData.iloc[:100, :]

fig = px.scatter_3d(df, x="world_rank", y="research", z="citations", title="3D Scatter Plot")
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))



fig.show()
```
