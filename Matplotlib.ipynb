## Two options for subplot

```python
#1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                          ncols = 2,
                                          figsize = (10 ,5))
ax1.plot(x, x);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(prices.keys(), prices.values())
ax4.hist(np.random.randn(100));
```

```python
# 2

fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
ax[0, 0].plot(x, np.random.random(1000));
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(prices.keys(), prices.values())
ax[1, 1].hist(np.random.randn(100));
```

# pyplot or matplotlib oo?

* basic => pyplot
*advanced => matplotlib oo

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
# pyplot
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target')
```

```python
# OO method mixed with pyplot

fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax = ax,
             xlim=(50, 79));
```

```python
# OO method

fig, ax = plt.subplots(figsize = (10, 6))

# plot the data

scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     )


# customize plot

ax.set(title='Heart Disease vs Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')

# add a legend

ax.legend(*scatter.legend_elements(), title='Target');

# add a horizontal line

ax.axhline(over_50['chol'].mean(),
           linestyle= '--')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# add data to ax0

scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     )


# customize ax0

ax0.set(title='Heart Disease vs Cholesterol Levels',
        ylabel='chol')

# add a legend to ax0

ax0.legend(*scatter.legend_elements(), title='Target');

# add a meanline 

ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--')

# add data to ax1


scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                     )


# customize ax1

ax1.set(title='Heart Disease vs Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate')

# add a legend to ax0

ax1.legend(*scatter.legend_elements(), title='Target');

# add a meanline 

ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--');

# add a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize= 16, fontweight = "bold");
```

```python
plt.style.use("seaborn-whitegrid")
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# add data to ax0

scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap="winter"
                     )


# customize ax0

ax0.set(title='Heart Disease vs Cholesterol Levels',
        ylabel='chol')

ax0.set_xlim([50, 80])


# add a legend to ax0

ax0.legend(*scatter.legend_elements(), title='Target');

# add a meanline 

ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--')

# add data to ax1


scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                     cmap="winter"
                     )
# customize ax1

ax1.set(title='Heart Disease vs Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate')

ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# add a legend to ax0

ax1.legend(*scatter.legend_elements(), title='Target');

# add a meanline 

ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--');

# add a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize= 16, fontweight = "bold");
```
